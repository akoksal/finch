```python
import io
import os
from google.cloud import vision
from google.cloud.vision import types
import matplotlib.pyplot as plt
```

```python
client = vision.ImageAnnotatorClient()
```

```python
file_name = os.path.join('Dataset/rec-5.jpg')

with io.open(file_name, 'rb') as image_file:
    content = image_file.read()

image = types.Image(content=content)
context = types.ImageContext(language_hints=["tr"])
response = client.text_detection(image=image, image_context=context)
```

```python
print(response.text_annotations[0].description)
```

```python
for text_block in response.text_annotations[1:]:
    if text_block.description in ["GAZ1", "ALISVERIS", "HIZM.", "TURZ", "GIDA", "SAN.", "VE"]:
        print(text_block)
```

```python
elements = []
for ta in response.text_annotations[1:]:
    my = 0
    for vertice in ta.bounding_poly.vertices:
        my += vertice.y
    mx = 0
    for vertice in ta.bounding_poly.vertices:
        mx += vertice.x
    mx/=4
    elements.append((ta.description, mx, my))
```

```python
elements = sorted(elements, key=lambda x: (x[2], x[1]))
desc = []
el =[]
for i in range(len(elements)-1):
    if elements[i+1][2] - elements[i][2] < 150:
        el.append(elements[i])
    else:
        desc.append(el)
        el = [elements[i]]
```

```python
for i in range(len(desc)):
    desc[i] = sorted(desc[i], key=lambda x: (x[1]))
```

```python
fin = ""
for d in desc:
    for l in d:
        fin += l[0]+" "
    fin += "\n"
print(fin)
```
