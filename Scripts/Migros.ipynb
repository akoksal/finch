```python
from selenium import webdriver
driver = webdriver.Chrome('/home/koksal/Desktop/finch/chromedriver')
```

```python
with open("cats.txt", "r") as f:
    links = list(set(f.read().splitlines()))
```

```python
def get_all_elements(baselink):
    page = 1
    iterlink = baselink+"?sayfa="+str(page)
    driver.get(iterlink)
    cat_elements = driver.find_elements_by_css_selector("ul.breadcrumb li")
    parent_cat = cat_elements[-2].text
    category = cat_elements[-1].text
    print("Parent category: "+parent_cat)
    print("Category: "+category)
    items = []
    while True:
        print(page)
        item_divs = driver.find_elements_by_css_selector("div.center.product-card-center")
        for item in item_divs:
            itemname = item.find_element_by_css_selector("h5.title.product-card-title").text
            itemprice = item.find_element_by_css_selector("div.price-tag").text
            #print(itemname,itemprice)
            items.append((itemname, itemprice))
        if len(item_divs)==0:
            break
        else:
            page+=1
            iterlink = baselink+"?sayfa="+str(page)
            driver.get(iterlink)
    return {"Ana kategori":parent_cat, "Kategori":category, "Item":items}
```

```python
elements = []
for link in links:
    element = get_all_elements(link)
    elements.append(element)
```

```python
import json
with open("items.json", 'w') as fp:
    json.dump(elements, fp)
```
